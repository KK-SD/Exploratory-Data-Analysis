# 6. Data Visualization for EDA <a class="anchor" id="four"></a>

Data visualization, using tools such as Matplotlib and Seaborn, is a fundamental aspect of the exploratory data analysis (EDA) process. By transforming abstract numerical data into clear and intuitive graphical representations, we gain a deeper understanding of the complexities and nuances of the dataset. Visualization provides an alternative, visual approach to analyzing data.

These visualizations can reveal patterns, trends, correlations, outliers, and the overall structure of the dataâ€”insights that may not be immediately obvious from raw data tables or basic statistics. They enable us to quickly comprehend the dataset, communicate findings effectively, and even tell compelling stories through data.

As discussed earlier in this notebook, Matplotlib and Seaborn will be the primary tools used to explore and visualize our data in greater detail.








## 6.1. Basic Plots

In data visualization, several foundational plot types help uncover patterns and trends in data. The most common and essential plot types include:

- Bar Plots: Used to compare categorical data.
- Line Plots: Ideal for showing trends or changes over time.
- Scatter Plots: Useful for analyzing relationships between two continuous variables.
- Histograms: Help visualize the distribution of a single variable.

Each plot type serves a unique purpose and is best suited for specific types of data and analyses.

### 6.1.1. Histogram Charts

Histograms are graphical representations of data frequencies, displayed as bars. They show how much of the dataset falls into specific, non-overlapping intervals called bins. The `.hist()` method in Matplotlib is used to create a simple histogram.

To plot a histogram using Matplotlib, you first need to import the library, then call the `.hist()` method on your dataset.



```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np  # Missing import for numpy

# Generate random data
data1 = np.random.randn(1000)  # A sample generation based on random

# Create histogram
plt.hist(data1)
plt.xlabel("x")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

### 6.1.2. Bar charts

Bar charts represent data with rectangular bars, either vertically or horizontally. The length of each bar is proportional to the value it represents. Bar charts are ideal for comparing two or more values, and the `.bar()` method in Matplotlib can be used to create them.

```python
# Data for bar plot
data2 = [5.0, 25.0, 50.0, 20.0]

# Create bar plot
plt.bar(range(len(data2)), data2)
plt.xlabel("x")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

### 6.1.3. Scatter Plot

Scatter plots display individual data points as dots or circles, showing relationships between two continuous variables. The `.scatter()` method in Matplotlib is used to create scatter plots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for plotting
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure with a custom size
plt.figure(figsize=(8, 6))

# Plot sine function with customized appearance
plt.plot(
    x, y, label='sin(x)', color='blue', linewidth=2, linestyle='-',
    marker='o', markersize=6, markerfacecolor='red', markeredgewidth=2,
    markeredgecolor='black'
)

# Title
plt.title('Sine Function')  # Sets the title of the plot

# Axis labels
plt.xlabel('x')
plt.ylabel('sin(x)')

# Customizing Tick Marks
plt.xticks(np.arange(0, 11, step=1))  # Set tick marks on the x-axis from 0 to 10
plt.yticks(np.arange(-1, 1.5, step=0.5))  # Set tick marks on the y-axis from -1 to 1.5

# Grid
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines with dashed lines and transparency

# Legend
plt.legend(loc='upper right')  # Add a legend at the upper right corner

# Save the figure as a PNG file
plt.savefig('sin_plot.png')

# Display the plot
plt.show()
```

### 6.1.4. Pie Charts

Pie charts are circular graphs divided into sectors (or wedges) that represent proportions of a whole. The `.pie()` method in Matplotlib is used to create pie charts. Each wedge is proportional to the value in the dataset, calculated as x/sum(X) for each value x in the input array X.


```python
import matplotlib.pyplot as plt

# Data for pie chart
labels = 'Health', 'Climate Change', 'Obesity'
sizes = [70.4, 20.6, 9.0]
colors = ['lightgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)

# Create pie chart
plt.pie(
    sizes, explode=explode, labels=labels, colors=colors, shadow=True,
    startangle=90
)

# Title
plt.title("What is the most concerning topic for the next 10 years? (100 people)")

# Legend
plt.legend(title="Answers")

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the plot
plt.show()
```

## 6.2. Data Visualization using Pandas package

The Pandas package includes several built-in tools for creating visualizations, making it a practical choice for exploring data, even though it is not a dedicated data visualization library. Using the `.plot()` function, you can generate various types of visualizations, such as line plots, bar plots, histograms, box plots, area plots, scatter plots, and pie charts, providing a quick and effective way to analyze your data.

### 6.2.1. Scatter Plot

The `DataFrame.plot.scatter()` function is used to create scatter plots. This function requires two arguments, typically the column names for the x and y axes. It enables labeling the axes using the column names to make the graph more informative.


```python
import pandas as pd

# Sample dataset
df = pd.DataFrame({
    'num_legs': [2, 4, 8, 0],
    'num_wings': [2, 0, 0, 0],
    'num_specimen_seen': [10, 2, 1, 8],
    'species': ['falcon', 'dog', 'spider', 'fish']
})

# Create a scatter plot between species and number of legs
df.plot.scatter(x='species', y='num_legs')
```

### 6.2.2. Pie Chart

The `DataFrame.plot.pie()` or `Series.plot.pie()` functions are used to generate pie charts. When applied to a DataFrame, it creates individual pie charts for each column, displaying them as subplots for a comprehensive view.

```python
import numpy as np
import pandas as pd

# Create a new dataset
my_df = pd.DataFrame(np.random.rand(6, 3), columns=('Column_1', 'Column_2', 'Column_3'))

# Plot pie charts for each column in the dataset
my_df.plot.pie(subplots=True, figsize=(15, 15))
```

### 6.2.3. Histogram

The `DataFrame.plot.hist()` function is used to create histograms, enabling the visualization of the distribution of numerical data in your dataset.

```python
# Plot histogram for the dataframe
df.plot.hist()
```

## 6.3. Subplots <a class="anchor" id="four-three"></a>

In data analysis, it is often useful to display multiple plots together for comparison or a broader understanding of the data. Pandas provides straightforward methods to create subplots, allowing multiple visualizations to be presented side by side or in a grid layout for holistic examination.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the empty canvas for the plots
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Two Subplots')

# Generate data for the plots
x = np.linspace(0, 100, 30)
y = np.random.rand(30)

# Plot the two scatter subplots
ax1.plot(x, y)
ax2.plot(x, -y)

# Display the plots
plt.show()
```

## 6.4. Additional Data Visualizations (Complex Plots) <a class="anchor" id="four-four"></a>

Beyond standard plots, statistical visualization methods provide deeper insights into datasets. This section highlights the use of Kernel Density Estimation (KDE) plots and correlation graphs, which can reveal intricate patterns and relationships among features.

### 6.4.1. Kernel Density Plot  <a class="anchor" id="four-four-one"></a>

[Kernel Density Estimation (KDE)](https://www.statsmodels.org/stable/examples/notebooks/generated/kernel_density.html) is a statistical method used to estimate the probability density function of a numerical variable. It leverages kernel functions to smooth and visualize the distribution of data, offering insights into its structure.


Key Benefits of KDE Plots:

- Range Insight: Quickly identifies the minimum, maximum, and mean values of a variable.
- Common Values: Highlights the most frequent values (peaks) in the dataset.
- Data Representativeness: Shows if the dataset is well-sampled across all value ranges.
- Classification Analysis: Demonstrates how a variableâ€™s distribution differs across classes.
- Distribution Transformation: Variables with heavy tails or non-Gaussian distributions may require transformations (e.g., logarithmic or power transforms) to improve model performance.

Plotting KDE with Seaborn:

For better readability and aesthetics, Seaborn is often preferred over Matplotlib for creating KDE plots. Below, we will demonstrate a KDE plot using the Iris dataset, which consists of four features: Sepal Length, Sepal Width, Petal Length, and Petal Width.

For further guidance on KDE parameters and how to properly represent the data, refer to the Seaborn documentation. [here](https://seaborn.pydata.org/generated/seaborn.kdeplot.html).

More information about the Iris dataset can be found [here](https://archive.ics.uci.edu/ml/datasets/iris).


```python
#let's use seaborn visualisation library
import seaborn as sns
```

```python
#load iris dataset
iris = sns.load_dataset("iris")
iris
```

```python
plt.figure(figsize=(20,5))#set figure size
ax = sns.kdeplot(data = iris) #plot definition
plt.title('KDE Plot Iris Dataset');  #set title name

#remove spines just to look better
sns.despine(top=True, right=True, left=False, bottom=False)
```

From the above plot, we can extract the following insights:
- Sepal Width (Blue Curve): Peaks around 3, indicating the mean and most common value. Values range between approximately 2 and 4.
- Petal Width (Red Curve): Shows two peaks, suggesting two common widths in the data.
- Petal Length (Green Curve): Bimodal shape with a wide range, indicating variability across the dataset.
- Sepal Length (Yellow Curve): Flat peak, showing all values are equally likely, with no dominant range.


### 6.4.2. Plotly Library  <a class="anchor" id="four-five"></a>

Plotly is a powerful library for creating interactive, publication-quality visualizations in Python. It supports a wide range of chart types, including line plots, scatter plots, bar charts, histograms, heatmaps, and more.

Key Features:
- Interactive Visualizations: Ideal for exploring data with hover, zoom, and filter capabilities.
- High-Level Interface: Plotly Express simplifies creating visualizations with minimal code.
- Customizable and Open-Source: Fully customizable and available for free on GitHub.

**Scatter Plot: <a class="anchor" id="four-five-one"></a>**
Using px.scatter, each data point is represented as a marker whose position corresponds to values from specified x and y columns.

- Bubble Charts: Add variable sizes to markers.
- Color and Hover Information: Include custom details using hover_data.
- Faceting: Create multiple subplots (small multiples) based on data categories, all sharing the same axes.

```python
# x and y given as DataFrame columns
import plotly.express as px
df = px.data.iris()
figure = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
figure.show()
```

```python
# Example: Scatter plots support faceting
import plotly.express as px
df = px.data.tips()
figure = px.scatter(df, x="total_bill", y="tip", color="smoker", facet_col="sex", facet_row="time")
figure.show()
```

**Statistical Charts<a class="anchor" id="four-five-two"></a>**

Plotly can create statistical visualizations like error bars, histograms, and tree plots.

Histograms:

- Represent data distribution by binning numerical values and showing counts per bin.
- Support aggregation functions like sum, average, and count.
- Can visualize data on categorical, date, or linear axes.

```python
import plotly.express as px

# Load the tips dataset from Plotly
df = px.data.tips()

# Create a histogram of the total bill
figure = px.histogram(df, x="total_bill")

# Display the histogram
figure.show()
```

Plotly histograms automatically group numerical or date data into bins for easy visualization. They can also handle raw categorical data, such as using the categorical variable "day" as the X-axis to display its frequency distribution.

```python
import plotly.express as px

# Load the tips dataset from Plotly
df = px.data.tips()

# Create a histogram of the 'day' column with custom category order
figure = px.histogram(
    df, x="day",
    category_orders=dict(day=["Thur", "Fri", "Sat", "Sun"])
)

# Display the histogram
figure.show()
```

# 7. Correlation and Covariance Analysis

Correlation and covariance are important statistical tools used to measure the relationship between two or more variables.

## 7.1. Correlation Analysis and Visualization  <a class="anchor" id="four-four-one"></a>

The correlation coefficient, often denoted as "r," measures the strength and direction of the linear relationship between two variables. One of the most common methods for calculating correlation is the [Pearson correlation coefficient](https://www.statisticshowto.com/probability-and-statistics/correlation-coefficient-formula/#Pearson). The resulting value ranges from +1 to -1, indicating the strength of the relationship between two variables.

- A positive correlation (r > 0) means that as one variable increases, the other also increases at a similar rate.
- A negative correlation (r < 0) means that as one variable increases, the other decreases.
- A correlation of r = 0 indicates no linear relationship between the variables.


**What can we learn from the Correlation Heatmap?**
- In the context of regression problems, identifying variables that are highly correlated (either positively or negatively) with the target variable can be valuable. This suggests that changes in the feature will likely lead to corresponding changes in the target.

- Correlation analysis can also help identify redundant features. If two variables are highly correlated, they may provide the same information, and you could choose to remove one to reduce data complexity.

**What we should not assume from the Correlation Heatmap?**

- A correlation coefficient close to +1 or -1 indicates a strong linear relationship between two variables, but values near 0 do not necessarily mean there is no relationship. It could mean the relationship is non-linear (e.g., exponential or quadratic). Visualizing the relationship through scatter plots can help confirm this.

- It's also crucial to remember that correlation does not imply causation. A high correlation between two variables does not mean one causes the other. For example, a study might find a strong correlation between time spent watching TV and aggressiveness in children. While this could suggest that watching TV influences aggression, the reverse could be true: perhaps more aggressive children watch more TV.

**Interpreting the Correlation Heatmap**

Correlation is often visualized using a heatmap, which displays the relationships between variables. The values along the diagonal will always be 1 since a variable is always perfectly correlated with itself. For analysis, we typically focus on one half of the heatmap to examine the relationships between different variables.



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset as a DataFrame
iris_data = load_iris(as_frame=True)
iris = iris_data.frame

# Drop the non-numeric column ('target' is the species column in sklearn's iris dataset)
numeric_iris = iris.drop(columns=["target"])

# Calculate the correlation matrix
corr = numeric_iris.corr()

# Plot the heatmap
ax = sns.heatmap(corr, linewidths=.5, annot=True, fmt=".2f", cbar=False)
plt.show()

```

```python
corr
```

From the above plot, we can extract the following insights:
* Sepal Length is strongly correlated with Petal Length (0.87) and Petal Width (0.82), but weakly correlated with Sepal Width (-0.12).
* Sepal Width is inversely correlated with Petal Length and Petal Width; as Sepal Width increases, Petal Length and Petal Width decrease.
* Petal Length and Petal Width are highly correlated (close to 1), meaning they tend to increase together.

## 7.2. Covariance

Covariance measures how two random variables vary together. It indicates the degree to which changes in one variable can be predicted by changes in another. The sign of the covariance reveals the direction of their linear relationship:

* Positive covariance: The variables tend to move in the same direction.
* Negative covariance: The variables tend to move in opposite directions.

The example code demonstrates how to calculate covariance using the iris dataset and visualize it with a Seaborn heatmap.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset from seaborn
iris = sns.load_dataset("iris")

# Exclude the non-numeric 'species' column
numeric_iris = iris.drop(columns=["species"])

# Compute the covariance matrix
covariance_matrix = numeric_iris.cov()

# Display the covariance matrix
print(covariance_matrix)

# Create a heatmap from the covariance matrix
plt.figure(figsize=(10, 8))
sns.heatmap(covariance_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Covariance Matrix Heatmap")
plt.show()
```
